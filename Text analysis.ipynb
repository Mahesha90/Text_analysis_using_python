```python
import re

data = """
Analyzes the teacher's lesson plan and prepares the lesson plan himself based on the initial task.\nPrepares an introduction to one method as the final thesis of the course.
Analytical Essay; practical work (short film or presentation or puppet); oral conversation.
Data collection
calculation
calculation
Assessment: written self-analysis based on the professional standard\n\nThe assessment consists of:\n•\a written self-analysis;\n•\a written presentation as a team work.\n\nPreparation of a written self-analysis based on the professional standard Teacher level 6 based on what you have learned so far and what you have experienced in practice. Preparing and delivering a presentation on the team's strengths and development areas.\n
asdasd
Bachelor thesis
Conducting a demo lesson in training - observation
Didactic analysis
As preliminary work, the student has found out the students' prior knowledge, level and learning motivation, and based on this, plans learning goals and learning activities, which the student presents when preparing lesson plans.
exam
Exam based on submitted homework: 50% group work, 50% individual homework
The exam consists of a test with open questions, an oral presentation and a written analytical review of sources, self-analysis through an interest education event in an open learning environment, and independent works of non-differentiated assessment.
Self analysis
Self-analysis work, aggregate analysis of the observed lessons, reflection of the conducted lessons, reflection after the class teacher's task, analysis of conducting an extracurricular event, self-analysis based on the professional standard, reflection tasks in practice seminars.
Self-development planning in terms of digital competences
Distinctive (A,B,C,D,E,F, mi)
The average grade of the submitted work
essay
Essay
Presentation
Evaluation method\n•\tAnalysis of the selected school's curriculum, evaluation procedure and general work plan.\n•\tPreparation of a class certificate with verbal evaluations.\n•\tIndependent work with literature. Preparing a presentation based on the literature read.\n\n
Assessment method: Analytical essay, multiple choice task and oral interview
Evaluation method: mini-survey
Written research projects will be evaluated.
Independent works make 50% of the final outcome.\nVideo to propose policy recommendations is 10% of the final outcome.\nExam is the analysis of the chosen theme with proposed policy recommendation is 40% of final outcome.
Independent work 1: topic mapping - diagnosis stage\n
Independent work 1: topic mapping - diagnosis stage\nIndependent work 2: thematic summary of sources
Independent work: accounting, i.e. drawing up an action research plan, describing the stages, working with literature
Independent work: preparation of three different educational and educational activity plans
Independent work: integration, notice and acknowledge, feedback.\nExam paper
Independent work: integration, notice and acknowledge, feedback; exam paper
Independent work: formulation of learning outcomes; choice of teaching methods with justification. Exam paper.
Independent work: analysis of a research article
Independent work: written analysis of a research article
Supervisor-teacher feedback.
Language level assignment
Written and oral reflection
Preparation of three different educational and educational activities plans.
Cooperation with the supervisor-teacher is collaborative, and the supervisor-teacher's feedback to the supervisee shows the trainee's readiness to take suggestions into account.\n\n
In cooperation with the supervisor-teacher of the training school, he finds out the subject knowledge, study skills and motivation of the students in the taught classes (feedback from the supervisor-teacher). When preparing lesson plans and study materials, the student takes into account the knowledge found about students' learning motivation, subject skills, etc. (lesson plans together with study materials are submitted to the lecturer). Students record video lessons and give feedback on each other's lessons.
The Western Plover
Reading task, memorization task, lesson outline analysis, exam.
Master's thesis
Master thesis research plan
Creating a music agenda, creating/editing a song play, presentation
Observation of music activities and conducting 4 consecutive music activities in two different age groups
Observation of musical activities by filling in a questionnaire supporting learner autonomy; planning music activities (creating an independent plan, consulting with the supervising teacher before performing the music activity; writing a short summary of self-reflection and feedback from the supervising teacher at the end of each music activity plan.
Teacher's work plan, integration, lesson plan, skill-building tasks in history and social studies class, analysis of study material, preparation of control tasks, exam
Teacher's work plan, integration, lesson plan, skill-building tasks, written exam
In the learning community, at least 8 materials in the STEAM field of study have been prepared in cooperation, which are ready to be published in eKoolikot
study folder
Study folder (see course program)
Tasks for creating a learning scenario
Adhering to the rules of safe, legal and ethical behavior when creating learning scenarios
Applying the principles of actively involving the student in learning scenarios and the Modern learning concept with various digital tools
Analysis of curricula, action plans, exam
Study material
Study material analysis, written exam at the end of the subject
A learning tool
Participation in group work - written and oral reflection
Piaget's experiments
Basics of teaching basic subjects - accounting\nBasic practice in grades 1-3 - exam
Practice study folder with video
One of the tasks of the internship is to conduct a well-thought-out lesson that uses group work (the student presents a well-thought-out lesson plan with a longer description of the course of the lesson). After the lesson, the student makes a brief analysis of the successful and unsuccessful moments of the lesson.
Practice folder
Practical tasks that require the application of theoretical knowledge
At least one collection of data on classroom practice with Prolearning and Observata app - data analysis
At least one collection of data on classroom practice with Prolearning and Observata app - data analysis
paper
reflection
reflection
Reflection, a log book created as a team effort
Reflections in homework
Reflections.
Reflection
Reflection, self-analysis
Group work, essay, creation of digital learning material
Same as last time.
Seminar task and homework, preparation of group work task, planning of support task.
Introductory self-analysis before the start of the school internship. Summary analysis of practice (including personal learning goals). Final seminar.
Oral self-analysis at the mid-term seminar and the final seminar; writing a short summary and self-reflection of the supervising teacher's feedback at the end of each music activity plan; planning and conducting music activities.
Feedback on the video watched by a fellow student, supervisor-teacher and lecturer's feedback on the course of practice.
Knowledge is tested in the exam of the subject of state law. The court and court proceedings course gives 15% of the grade of the subject.
Planning the action plan, getting to know the practices and requirements of the internship school under the supervision of the head of education of the internship school or supervisor-teacher. Planning lesson plans.
Action plans
Action plans, exam
Action plans, augmented reality game
Action plans, creation of a video guide for learning activities integrated with robotics.
Action plans, video guide
Tinnukva
Timetables
Preparation of lesson plans. Preparation of an action plan.
Preparation of lesson plans. Conducting lessons. Lesson analysis.
Lesson outlines, summaries of conducted lessons, class teacher's task and extracurricular event analyses.
Lesson outlines, summaries of conducted lessons.
Analysis of one master's thesis\nAnalysis of theoretical sources\nMusatnd of the theoretical part of the master's thesis
Analysis of one master's thesis
A presentation to peers introducing one of the study sets
The tasks are formulated linguistically and formally correctly.
Together with the lesson plans, the student submits excerpts/solutions etc. from the educational materials selected for the lesson and also adds materials created by him (including lesson tests and tests).
The student proves his skills through the presented relevant lesson plans and study materials. In addition, video clips from one selected lesson.
An internship action plan drawn up by the student and coordinated with the internship supervisor, \nwhich takes into account the study content and learning outcomes of the informatics curriculum.
Completion of the written theory part of the research paper and its oral presentation.
Analyzes of observed and conducted lessons, self-analyses, lesson plans, feedback from instructors
Analysis of the observed lessons, reflection of the conducted lessons, reflections on the task of leading the class and an extracurricular event.
Completing observation sheets, conducting an event or lesson with the class, writing an analysis of practice
Presentation of analyzes of observational practice
Conducting at least 8 robomathematics lessons as part of a demo lesson and a credit lesson/interest circle - written and oral reflection
Conducting at least 8 STEAM classes within a subject class or interest group - written and oral reflection\n
Collecting and analyzing data for at least one lesson - written and oral reflection
Application of the TTS form in at least one lesson - written and oral reflection
Being the main author of the technical part of at least 1 learning assignment - a completed H5P-based assignment
From the gaps
From the gaps
•\fulfilling the tasks on the observation sheets;\n•\tpreparing and conducting the lessons;\n•\writing a personal analysis of the practical experience\n
"""

# Define patterns for different skills
analytical_thinking_pattern = re.compile(r'\b(Analysis|Analytical|Evaluation)\b', re.IGNORECASE)
communication_pattern = re.compile(r'\b(Presentation|Oral|Communication)\b', re.IGNORECASE)
research_writing_pattern = re.compile(r'\b(Written|Research|Essay)\b', re.IGNORECASE)
problem_solving_pattern = re.compile(r'\b(Conducting|Practical|Application)\b', re.IGNORECASE)
teamwork_collaboration_pattern = re.compile(r'\b(Group|Cooperation|Teamwork)\b', re.IGNORECASE)
digital_competence_pattern = re.compile(r'\b(Digital)\b', re.IGNORECASE)
educational_planning_pattern = re.compile(r'\b(Lesson|Study|Teacher|Planning)\b', re.IGNORECASE)

# Initialize dictionaries to store skills
skills = {
    "Analytical and Critical Thinking Skills": set(),
    "Communication and Presentation Skills": set(),
    "Research and Writing Skills": set(),
    "Problem Solving and Practical Skills": set(),
    "Teamwork and Collaboration": set(),
    "Digital Competence": set(),
    "Educational Planning and Instructional Design": set(),
}

# Split data into lines
lines = data.split('\n')

# Iterate through lines and identify skills
for line in lines:
    for skill_category, pattern in zip(skills.keys(), [analytical_thinking_pattern, communication_pattern,
                                                      research_writing_pattern, problem_solving_pattern,
                                                      teamwork_collaboration_pattern, digital_competence_pattern,
                                                      educational_planning_pattern]):
        if pattern.search(line):
            skills[skill_category].add(line.strip())

# Print the identified skills for each category
for skill_category, skill_set in skills.items():
    print(f"{skill_category}:\n", "\n".join(skill_set))
    print("\n" + "="*50 + "\n")
```
